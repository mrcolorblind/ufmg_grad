```python
import io
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
conn = sqlite3.connect('despesas.db')
cursor = conn.cursor()
```

```python
f = io.open('despesas.sql', 'r', encoding='UTF-8')
sql = f.read()
cursor.executescript(sql)
```

```python
ANS = []
```

```python
## Seção 4.1
questao = 1

query = """
SELECT COD_SUBFUNCAO, NOME_SUBFUNCAO
FROM SUBFUNCAO
WHERE COD_SUBFUNCAO IN
(SELECT COD_SUBFUNCAO
FROM DETALHAMENTO_FUNCAO
WHERE COD_FUNCAO =
(SELECT COD_FUNCAO
FROM DETALHAMENTO_FUNCAO
GROUP BY COD_FUNCAO
ORDER BY COUNT(*) DESC LIMIT 1))
"""

df = pd.read_sql_query(query, conn)
display(df)
ANS.append([questao, len(df), len(df.columns), query])
```

```python
## Seção 4.1
questao = 2
query = """
SELECT NOME_FUNCAO, NOME_SUBFUNCAO, VALOR_DESPESA
FROM DESPESA_PUBLICA AS DP
INNER JOIN DETALHAMENTO_FUNCAO AS DF
ON DF.COD_DETALHAMENTO = DP.COD_DETALHAMENTO
INNER JOIN SUBFUNCAO AS S
ON S.COD_SUBFUNCAO = DF.COD_SUBFUNCAO
INNER JOIN FUNCAO AS F
ON F.COD_FUNCAO = DF.COD_FUNCAO
WHERE VALOR_DESPESA > 200000 
GROUP BY DP.COD_DETALHAMENTO
"""

df = pd.read_sql_query(query, conn)
display(df)
ANS.append([questao, len(df), len(df.columns), query])
```

```python
## Seção 4.1
questao = 3
query = """
SELECT D.COD_ORGAO, O.NOME_ORGAO, SUM(VALOR_DESPESA)
FROM DESPESA_PUBLICA AS D
INNER JOIN ORGAO AS O
ON O.COD_ORGAO = D.COD_ORGAO
GROUP BY O.COD_ORGAO
"""

df = pd.read_sql_query(query, conn)
display(df)
ANS.append([questao, len(df), len(df.columns), query])
```

```python
## Seção 4.1
questao = 6

query = """
SELECT NOME_FUNCAO, NOME_SUBFUNCAO, SUM(VALOR_DESPESA)
FROM DETALHAMENTO_FUNCAO AS DF
INNER JOIN DESPESA_PUBLICA AS DP
ON DP.COD_DETALHAMENTO = DF.COD_DETALHAMENTO
INNER JOIN FUNCAO AS F
ON DF.COD_FUNCAO = F.COD_FUNCAO
INNER JOIN SUBFUNCAO AS S
ON DF.COD_SUBFUNCAO = S.COD_SUBFUNCAO
WHERE DP.COD_DETALHAMENTO IN
(SELECT COD_DETALHAMENTO
FROM DESPESA_PUBLICA AS DP
WHERE MES_DESPESA BETWEEN 7 AND 9
AND ANO_DESPESA = 2018
AND COD_PROGRAMA_GOVERNO =
(SELECT COD_PROGRAMA_GOVERNO
FROM PROGRAMA_GOVERNO
WHERE NOME_PROGRAMA_GOVERNO = "MAIS MEDICOS"))
GROUP BY DP.COD_DETALHAMENTO
"""

df = pd.read_sql_query(query, conn)
display(df)
ANS.append([questao, len(df), len(df.columns), query])
```

```python
## Seção 4.1
questao = 7

query = """
SELECT NOME_ORGAO,
COUNT(DISTINCT COD_PROGRAMA_GOVERNO) AS NUMERO_DE_PROGRAMAS
FROM ORGAO AS O
INNER JOIN DESPESA_PUBLICA AS DP
ON O.COD_ORGAO = DP.COD_ORGAO
WHERE DP.VALOR_DESPESA > 0
GROUP BY NOME_ORGAO
"""

df = pd.read_sql_query(query, conn)
display(df)
ANS.append([questao, len(df), len(df.columns), query])
```

```python
## Seção 4.1
questao = 8

query = """
SELECT NOME_PROGRAMA_GOVERNO, NOME_SUBFUNCAO, SOMA
FROM 
(SELECT NOME_PROGRAMA_GOVERNO, COD_DETALHAMENTO, SUM(VALOR_DESPESA) AS SOMA
FROM DESPESA_PUBLICA AS DP
INNER JOIN PROGRAMA_GOVERNO AS P
ON P.COD_PROGRAMA_GOVERNO = DP.COD_PROGRAMA_GOVERNO
GROUP BY P.COD_PROGRAMA_GOVERNO) AS P
INNER JOIN DETALHAMENTO_FUNCAO AS DF
ON DF.COD_DETALHAMENTO = P.COD_DETALHAMENTO
INNER JOIN SUBFUNCAO AS S
ON S.COD_SUBFUNCAO = DF.COD_SUBFUNCAO
"""

df = pd.read_sql_query(query, conn)
display(df)
ANS.append([questao, len(df), len(df.columns), query])
```
